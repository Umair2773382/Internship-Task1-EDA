```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
df = pd.read_csv('titanic/train.csv')
# Display the first few rows
df.head()
```

```python
# Checking for missing values in each column 
df.isnull().sum()

# Fill missing numeric values with median
df['Age'].fillna(df['Age'].median(), inplace=True) 

# Fill missing categorical values with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

#checkig Duplicates
df.duplicated().sum()

# Remove duplicates
df.drop_duplicates(inplace=True)

# Check the cleaned data
df.info()
```

```python
# Bar chart for a categorical variable (e.g., 'Sex')
sns.countplot(x='Sex', data=df)
plt.title('Count of Passengers by Sex')
plt.show()

# Histogram for a numeric variable (e.g., 'Age')
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Summary of Insights

1. The number of male passengers is higher than female passengers.
2. Most passengers are in the age group of 2040 years.
3. There is a moderate positive correlation between Fare and Pclass.
4. Missing values were mostly found in the 'Age' and 'Embarked' columns and have been handled.
